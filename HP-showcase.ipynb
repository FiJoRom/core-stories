```python
from spacy.en import English
nlp = English()
```

```python
f = open("../data/kindle/kindle-015/text/1e03e4aa6532f8833a0eefc64a9eb3c1.txt","r")
doc = nlp(f.read())
f.close()
```

```python
len([x for x in doc.sents])
```

```python
sent_happs = np.zeros(len([x for x in doc.sents]))
```

```python
for i,sent in enumerate(doc.sents):
    sent_happs[i] = my_LabMT.score(dictify([str(x) for x in sent]))
```

```python
plt.figure(figsize=(18,6))
plt.scatter(np.arange(len(sent_happs)),sent_happs,s=6,edgecolor="k",facecolor="k")
plt.xlim([0,len(sent_happs)])
plt.xlabel("Sentence")
plt.ylabel("Happs")
plt.savefig("HP-sentences.png",bbox_inches="tight",dpi=600)
```
