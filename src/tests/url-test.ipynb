```python
x = range(5)
y = range(5)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, "r-", label=r"\url{http://google.com}")

ax.legend()

fig.savefig("test-url.pdf",bbox_inches="tight")
call("open test-url.pdf",shell=True)
```

```python
x = range(5)
y = range(5)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, "r-", label=r"\href{http://google.com}")

ax.legend()

fig.savefig("test-href.pdf",bbox_inches="tight")
call("open test-href.pdf",shell=True)
```

```python
x = range(5)
y = range(5)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, "r-", label=r"{\href{http://google.com}{link}}")

ax.legend()

fig.savefig("test-href-link.pdf",bbox_inches="tight")
call("open test-href-link.pdf",shell=True)
```
